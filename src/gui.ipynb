```python
import tkinter as tk
import math

class Node:
    def __init__(self, canvas, x, y, text, shape):
        self.canvas = canvas
        self.x = x
        self.y = y
        self.text = text
        self.shape = shape
        self.text_id = self.canvas.create_text(self.x, self.y, text=self.text, fill="black")
        bbox = self.canvas.bbox(self.text_id)
        if self.shape == "circle":
            r = max(bbox[2]-bbox[0], bbox[3]-bbox[1]) // 2 + 10
            self.id = self.canvas.create_oval(self.x-r, self.y-r, self.x+r, self.y+r)
        elif self.shape == "rectangle":
            self.id = self.canvas.create_rectangle(bbox[0]-10, bbox[1]-10, bbox[2]+10, bbox[3]+10)
        #self.text_id = self.canvas.create_text(self.x, self.y, text=self.text, fill="black")
        self.canvas.tag_bind(self.id, '<Button-3>', self.click)
        self.canvas.tag_bind(self.text_id, '<Button-3>', self.click)
        self.canvas.tag_bind(self.id, '<B1-Motion>', self.move)
        self.canvas.tag_bind(self.text_id, '<B1-Motion>', self.move)

    def move(self, event):
        dx = event.x - self.x
        dy = event.y - self.y
        self.canvas.move(self.id, dx, dy)
        self.canvas.move(self.text_id, dx, dy)
        self.x = event.x
        self.y = event.y
        self.canvas.update()
        

    def click(self, event):
        print(self.text)

class Arrow:
    def __init__(self, canvas, node1, node2):
        self.canvas = canvas
        self.node1 = node1
        self.node2 = node2
        self.id = self.canvas.create_line(self.node1.x, self.node1.y, self.node2.x, self.node2.y, arrow=tk.LAST)

    def update(self):
        dx = self.node2.x - self.node1.x
        dy = self.node2.y - self.node1.y
        angle = math.atan2(dy, dx)
        r = 20
        x1 = self.node1.x + r * math.cos(angle)
        y1 = self.node1.y + r * math.sin(angle)
        x2 = self.node2.x - r * math.cos(angle)
        y2 = self.node2.y - r * math.sin(angle)
        self.canvas.coords(self.id, x1, y1, x2, y2)

class App:
    def __init__(self, places, transitions, arcs):
        self.root = tk.Tk()
        self.canvas = tk.Canvas(self.root, width=500, height=500)
        self.canvas.pack()#
        self.places = places
        for place in self.places:
            self.places[place]["gui_object"] = Node(self.canvas, 100, 100, place, "circle")
        
        self.transitions = transitions
        for trans in self.transitions:
            self.transitions[trans]["gui_object"] = Node(self.canvas, 100, 100, trans, "rectangle")

        for
        self.arrow = Arrow(self.canvas, self.node1, self.node2)
        self.root.after(100, self.update)

        self.root.mainloop()

    def update(self):
        self.arrow.update()
        self.root.after(100, self.update)



```
